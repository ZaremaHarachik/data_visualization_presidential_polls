## Drop unnecessary columns and rows

### 1. Drop columns with only one value 

```python
polls = polls.drop(columns=['cycle', 'branch','matchup', 'forecastdate'])
```

### 2. Keep only the rows with "now-cast" in the type column and drop that column

```python
polls = polls.query('type == "now-cast"').copy()
polls = polls.drop(columns='type')
```

### 3. Keep only the rows that have 'lv' or 'rv' in the population column

In this case study I need just likely voters ("lv") and registered voters("rv")

```python
polls.query('population != "lv" & population != "rv"')   
polls = polls.query('population in ["lv","rv"]').copy()
```

### 4. Find unnecessary index columns and then drop them

```python
polls.loc[:, polls.nunique() == polls.shape[0]]  #display columns
polls = polls.drop(columns=['poll_id', 'question_id'])
```

### 5. Find columns that start with 'adjpoll' and then drop them

```python
polls.loc[:, polls.columns.str.startswith('adjpoll')]
polls = polls.loc[:, ~polls.columns.str.startswith('adjpoll')]
```

### 6. Drop the columns for Johnson and McMullen

```python
polls = polls.drop(columns=['rawpoll_johnson', 'rawpoll_mcmullin'])
```

### 7. Check the non-null rows in the multiversions column and drop that column

```python
polls.loc[polls.multiversions.notnull()]
polls = polls.drop(columns='multiversions')
```

### 8. Drop 3 more columns

```python
polls = polls.drop(columns=['createddate','timestamp','url'])
polls.head()
```

## Rename voting columns

```python
polls = polls.rename(columns={'rawpoll_clinton':'clinton_pct',
                              'rawpoll_trump':'trump_pct'})
```

## Fix object columns that should be numeric or datetime objects

```python
polls.select_dtypes('object').head(2)
```

### Convert the object columns that are dates to datetime objects

```python
date_cols = ['startdate', 'enddate']
polls[date_cols] = polls[date_cols].apply(pd.to_datetime)
```

This block of code starts by listing the unique values in the state column. That uncovers the fact that some state names
have suffixes like CD-1 or CD-2 where CD refers to Congressional District. So the next statement list all the state values

# Prepare the data (third phase)

## Add the columns for grouping and filtering

### 1. Add a voter_type column and drop the population column

```python
polls['voter_type'] = polls.population.replace(
    {'lv':'likely','rv':'registered'})
polls = polls.drop(columns='population')
```

### 2. Add a gap column for the gap between the Clinton and Trump percents

```python
polls['gap'] = polls.clinton_pct - polls.trump_pct
```

### 3. Add a state_gap summary column that contains the gap for each state

```python
polls['state_gap'] = polls.groupby('state').gap.transform(func='mean')
```

### 4. Add a swing column that identifies states with a less han a 7 percent gap

```python
polls['swing'] = (polls.state != "U.S.") & (abs(polls.state_gap) < 7)
```

#### The resulting DataFrame

```python
polls.head()
```

### Display the rows for swing states with query() method

```python
polls.query('swing == True')
```

## Create a new DataFrame in long form

### Melt the data to add two columns

```python
polls_long = polls.melt(
    id_vars=['state', 'enddate', 'voter_type', 'state_gap', 'swing'],
    value_vars=['clinton_pct', 'trump_pct'],
    var_name='candidate',
    value_name='percent')
```

### Clean up the strings in the candidate column

```python
polls_long['candidate'] = polls_long.candidate.str.replace('_pct', '').str.title()
```

## Take an early plot of the long data with Seaborn

```python
colors = ['#2281c4', '#d75c5d']             #blue and red colors
sns.set_palette(sns.color_palette(colors))  #change color palette

sns.lineplot(data=polls_long, x='enddate', y='percent',
             hue='candidate', errorbar=None)
```

## Add monthly bins to the DataFrame

```python
import datetime as dt
dates = pd.date_range('10/01/2015', '12/01/2016', freq='ME')
bin_labels = [dt.datetime.strftime(x, '%b %Y') for x in dates[1:]]
monthly_bins = pd.cut(x=polls_long.enddate, bins=dates, labels=bin_labels)
polls_long['month_bin'] = monthly_bins
polls_long.head()
```

## Add an average percent column for each month

```python
polls_long['month_pct_avg'] = polls_long.groupby(
    ['candidate', 'state', 'month_bin'], observed=True).percent.transform(func='mean')
polls_long.head()
```

## Save the wide and long DataFrames as pickle file

```python
polls.to_pickle('polls_prepared.pkl')
polls_long.to_pickle('polls_long.pkl')
```

# Analyze and Visualize the data (fourth and fifth phases) 

For better understaending the relationships between the data items I will show how to use 
plots to analyze the trends in the polling data. Here, the first plot is for the national polls, 
and the second one is for the swing states. If we look at the plots, we can see the differences in the 
results. This shows that Clinton was leading in the national polls in the early month, but these polls 
tightened as the election date neared. Similarly, Clinton was leading in the early polls for just the 
swing states, but these polls became extremely close as the election date neared with Trump taking a brief
ead in August.

## Plot the national and swing state polls.

### The national polls

```python
ax = sns.lineplot(data=polls_long.query('state == "U.S."'),
        x = 'month_bin', y = 'month_pct_avg', hue='candidate', errorbar=None)

ax.set(title='U.S. Polls', xlabel='Month', ylabel='Average Percent')
ax.tick_params('x', labelrotation=45)
```

### The swing state polls

```python
ax = sns.lineplot(data=polls_long.query('swing == True'),
         x='month_bin', y='month_pct_avg', hue='candidate', errorbar=None)

ax.set(title='Swing State Polls', xlabel='Month', ylabel='Average Percent')
ax.tick_params('x', labelrotation=45)
```
